## Demographics of Sea Otters in the Santa Barbara Channel

![](https://prd-wret.s3-us-west-2.amazonaws.com/assets/palladium/production/s3fs-public/styles/side_image/public/thumbnails/image/Sea_Otter_mom_and_pup_MG_3407.jpg)

Source: https://www.sciencebase.gov/catalog/item/585c5ac9e4b01224f329bc0a

```python
import pandas as pd
```

```python
raw_sea_otter_df = pd.read_csv('https://www.sciencebase.gov/catalog/file/get/585c5ac9e4b01224f329bc0a?f=__disk__22%2Ffe%2Fd8%2F22fed85e6822cf2a4117cb6b740bb6c2ae648773')
raw_sea_otter_df.head(5)
```

#### Show general statistics about the dataset -- don't worry about transformations yet

```python
raw_sea_otter_df.info()
```

```python
raw_sea_otter_df.describe()
```

#### How many sea otters are included in the observations?

```python
len(raw_sea_otter_df['Otter Number'].unique())
```

#### What is the gender distribution?

```python
raw_sea_otter_df['Sex'].value_counts()
```

```python
raw_sea_otter_df['Sex'].value_counts().plot(kind='barh',
                                            title='Sea Otter Genders')
```

#### What is the average age of the sea otters studied 

```python
raw_sea_otter_df['Age'].mean()
```

```python
round(raw_sea_otter_df['Age'].mean(), ndigits=2)
```

#### Where are they normally found

```python
raw_sea_otter_df['General Location'].value_counts().plot(kind='barh',
                                                         title='General Location')
```

#### What are their observed weights?

```python
raw_sea_otter_df['Weight (kg)'].nlargest()
```

```python
raw_sea_otter_df['Weight (kg)'].nsmallest(3)
```

#### Which sea otters are alive?

```python
raw_sea_otter_df[raw_sea_otter_df['Outcome'] == 'Survived']
```

#### Which sea otters are alive and male

```python
raw_sea_otter_df.dtypes
```

```python
raw_sea_otter_df[(raw_sea_otter_df['Outcome'] == 'Survived') & 
                 (raw_sea_otter_df['Sex'] == 'm')]
```

#### How many otters were captured by year?

```python
raw_sea_otter_df['Capture Date'] = pd.to_datetime(raw_sea_otter_df['Capture Date'])
raw_sea_otter_df.dtypes
```

```python
raw_sea_otter_df['Capture Date'].dt.year.value_counts()
```
